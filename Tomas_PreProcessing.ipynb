```python
path_to_folder = "/Users/tomashegewisch/Downloads/data_set/train_elephants"
image_paths = os.listdir(path_to_folder)
```

```python
plt.figure(figsize=(25,25))
for i in range(5):
    file = np.random.choice(image_paths)
    image_path= os.path.join(path_to_folder, file)
    img=mpimg.imread(image_path)
    ax=plt.subplot(1,5,i+1)
    ax.title.set_text(file)
    plt.imshow(img)
```

```python
img_data_array=[]
class_name=[]
IMG_WIDTH=330
IMG_HEIGHT=330

image_path="/Users/tomashegewisch/Downloads/data_set/train_elephants"
image_paths = os.listdir(image_path)
name_of_animal = 'elephant'
for img_name in tqdm(image_paths):
    image=cv2.imread(image_path+'/'+img_name, cv2.COLOR_RGB2GRAY)
    #image=cv2.imread(image_path+'/'+img_name, cv2.COLOR_BGR2RGB)
    image=cv2.resize(image, (IMG_HEIGHT, IMG_WIDTH),interpolation = cv2.INTER_AREA)
    image=np.array(image)
    image=image.astype('float32')
    image/= 255 
    img_data_array.append(image)
    class_name.append(name_of_animal)
```
